```python
import cv2
import numpy as np
from ultralytics import YOLO
from tqdm import tqdm

# === CONFIGURATION ===
VIDEO_PATH = "/content/shortened3min_vid.mp4"
YOLO_SEG_MODEL_PATH = "/content/bestyoloseg.pt"
OUTPUT_PATH = "/content/segmented_clean.mp4"

# === LOAD MODEL ===
model = YOLO(YOLO_SEG_MODEL_PATH)

# === VIDEO SETUP ===
cap = cv2.VideoCapture(VIDEO_PATH)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

out = cv2.VideoWriter(OUTPUT_PATH, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# === COLORS FOR CLASSES ===
np.random.seed(42)
class_colors = {}

# === PROCESS VIDEO ===
print("🎥 Processing video (masks only)...")
for _ in tqdm(range(frame_count), desc="Processing"):
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame, verbose=False)[0]
    blended = frame.copy()
    labels = set()

    if results.masks is not None:
        for mask_tensor, cls_tensor in zip(results.masks.data, results.boxes.cls):
            mask = mask_tensor.cpu().numpy()
            cls = int(cls_tensor.item())
            labels.add(model.names[cls])

            if cls not in class_colors:
                class_colors[cls] = np.random.randint(0, 255, size=3).tolist()
            color = class_colors[cls]

            binary_mask = (mask > 0.5).astype(np.uint8)
            binary_mask = cv2.resize(binary_mask, (width, height), interpolation=cv2.INTER_NEAREST)

            colored_mask = np.zeros_like(frame, dtype=np.uint8)
            for i in range(3):
                colored_mask[:, :, i] = binary_mask * color[i]

            blended = cv2.addWeighted(blended, 1.0, colored_mask, 0.4, 0)


    # Draw labels at top-left
    y = 30
    for label in sorted(labels):
        cv2.putText(blended, label, (10, y), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        y += 30

    out.write(blended)

cap.release()
out.release()
print("✅ Done! Clean segmentation video saved to:", OUTPUT_PATH)
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
from tqdm import tqdm

# === CONFIGURATION ===
VIDEO_PATH = "/content/shortened3min_vid.mp4"           # Path to your input video
YOLO_SEG_MODEL_PATH = "/content/bestyoloseg.pt"         # Path to your trained segmentation model
YOLO_DET_MODEL_PATH = "/content/yolov8obj.pt"      # Path to your YOLOv8 Nano object detection model
OUTPUT_PATH = "/content/segplusobj.mp4" # Output path
MAX_FRAMES = 300  # Limit processing to first 300 frames

# === FIXED CLASS COLORS (BGR for OpenCV) ===
class_colors = {
    0: [0, 0, 0],         # Black Background
    1: [255, 200, 0],     # Abdominal Wall
    2: [0, 100, 255],     # Liver
    3: [0, 255, 0],       # Gastrointestinal Tract
    4: [0, 255, 255],     # Fat
    5: [255, 0, 0],       # Grasper
    6: [180, 105, 255],   # Connective Tissue
    7: [0, 0, 255],       # Blood
    8: [255, 255, 0],     # Cystic Duct
    9: [255, 0, 255],     # L-hook Electrocautery
    10: [160, 255, 160],  # Gallbladder
    11: [100, 0, 255],    # Hepatic Vein
    12: [200, 200, 200],  # Liver Ligament
}

# === LOAD BOTH MODELS ===
# Load the segmentation model
seg_model = YOLO(YOLO_SEG_MODEL_PATH)

# Load the object detection model (YOLOv8 Nano)
det_model = YOLO(YOLO_DET_MODEL_PATH)

# === VIDEO SETUP ===
cap = cv2.VideoCapture(VIDEO_PATH)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Output video writer
out = cv2.VideoWriter(OUTPUT_PATH, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# === PROCESS VIDEO (LIMITED TO FIRST 300 FRAMES) ===
print("🎥 Processing first 300 frames with segmentation and detection...")
for _ in tqdm(range(min(MAX_FRAMES, frame_count)), desc="Processing"):
    ret, frame = cap.read()
    if not ret:
        break

    # 1. Apply segmentation mask (from YOLOv8 segmentation model)
    seg_results = seg_model(frame, verbose=False)[0]
    blended = frame.copy()
    labels = set()

    if seg_results.masks is not None:
        masks = seg_results.masks.data.cpu().numpy()  # shape: [N, h, w]
        masks_resized = [cv2.resize((m > 0.5).astype(np.uint8), (width, height), interpolation=cv2.INTER_NEAREST)
                         for m in masks]
        classes = seg_results.boxes.cls.cpu().numpy().astype(int)

        for binary_mask, cls in zip(masks_resized, classes):
            labels.add(cls)  # Use the class number directly
            color = class_colors.get(cls, [128, 128, 128])  # fallback: gray

            # Create colored mask
            colored_mask = np.stack([
                binary_mask * color[0],
                binary_mask * color[1],
                binary_mask * color[2]
            ], axis=-1).astype(np.uint8)

            blended = cv2.addWeighted(blended, 1.0, colored_mask, 0.4, 0)

    # 2. Apply bounding boxes (from YOLOv8 Nano detection model)
    det_results = det_model(frame, verbose=False)[0]
    if det_results.boxes is not None:
        for box in det_results.boxes.data.cpu().numpy():
            x1, y1, x2, y2, conf, cls = box
            cls = int(cls)
            color = class_colors.get(cls, [255, 255, 255])  # Default white box

            # Draw bounding box
            cv2.rectangle(blended, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)

            # Label text
            label = f"{det_model.names[cls]} {conf:.2f}"
            cv2.putText(blended, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # === Add class labels at the top ===
    y_offset = 30
    for cls in sorted(labels):
        label_text = seg_model.names[cls] if isinstance(seg_model.names, dict) else str(cls)
        color = class_colors.get(cls, [128, 128, 128])
        cv2.putText(blended, label_text, (10, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
        y_offset += 30

    # Write the frame with segmentation masks and bounding boxes
    out.write(blended)

cap.release()
out.release()
print(f"✅ Done! Processed first 300 frames. Segmented and detected video saved to: {OUTPUT_PATH}")
```
